```python
n_blizzard = []
s_blizzard = []
e_blizzard = []
w_blizzard = []

valley = []

entrance = ()
last = ()

x_max = None
y_max = None

try:
    f = open("input", encoding = 'utf-8')
    y = 0
    for row in f:
        if entrance == ():
            x_entrance = row.index('.')
            entrance = (x_entrance, y)
        elif entrance != () and row.count('#') == len(row) - 1:
            x_exit = row.index('.')
            last = (x_exit, y)
        else:
            row = row.strip()
            x_max = len(row) - 1
            for i in range(len(row)):
                if row[i] == '>':
                    e_blizzard.append([i, y])
                if row[i] == '<':
                    w_blizzard.append([i, y])
                if row[i] == 'v':
                    s_blizzard.append([i, y])
                if row[i] == '^':
                    n_blizzard.append([i, y])

        y += 1
    y_max = y - 1

finally:
    f.close()

positions = set()
positions.add(entrance)

found_last = False

minute = 0
while True:
    minute += 1
    for blizzard in e_blizzard:
        blizzard[0] += 1
        if blizzard[0] == x_max:
            blizzard[0] = 1
    for blizzard in w_blizzard:
        blizzard[0] -= 1
        if blizzard[0] == 0:
            blizzard[0] = x_max - 1
    for blizzard in s_blizzard:
        blizzard[1] += 1
        if blizzard[1] == y_max:
            blizzard[1] = 1
    for blizzard in n_blizzard:
        blizzard[1] -= 1
        if blizzard[1] == 0:
            blizzard[1] = y_max - 1
    
    new_positions = set()
    # print("Current pos", positions)
    for pos in positions:
        # print(pos)
        if [pos[0], pos[1]] not in e_blizzard and \
            [pos[0], pos[1]] not in w_blizzard and \
            [pos[0], pos[1]] not in s_blizzard and \
            [pos[0], pos[1]] not in n_blizzard:
            new_positions.add(pos)

        if 0 < pos[0] + 1 and pos[0] + 1 < x_max and 0 < pos[1] and pos[1] < y_max and \
            [pos[0] + 1, pos[1]] not in e_blizzard and \
            [pos[0] + 1, pos[1]] not in w_blizzard and \
            [pos[0] + 1, pos[1]] not in s_blizzard and \
            [pos[0] + 1, pos[1]] not in n_blizzard:
            new_positions.add((pos[0] + 1, pos[1]))

        if 0 < pos[0] and pos[0] < x_max and 0 < pos[1] + 1 and pos[1] + 1 < y_max and \
            [pos[0], pos[1] + 1] not in e_blizzard and \
            [pos[0], pos[1] + 1] not in w_blizzard and \
            [pos[0], pos[1] + 1] not in s_blizzard and \
            [pos[0], pos[1] + 1] not in n_blizzard:
            new_positions.add((pos[0], pos[1] + 1))

        if 0 < pos[0] - 1 and pos[0] - 1 < x_max and 0 < pos[1] and pos[1] < y_max and \
            [pos[0] - 1, pos[1]] not in e_blizzard and \
            [pos[0] - 1, pos[1]] not in w_blizzard and \
            [pos[0] - 1, pos[1]] not in s_blizzard and \
            [pos[0] - 1, pos[1]] not in n_blizzard:
            new_positions.add((pos[0] - 1, pos[1]))

        if 0 < pos[0] and pos[0] < x_max and 0 < pos[1] - 1 and pos[1] - 1 < y_max and \
            [pos[0], pos[1] - 1] not in e_blizzard and \
            [pos[0], pos[1] - 1] not in w_blizzard and \
            [pos[0], pos[1] - 1] not in s_blizzard and \
            [pos[0], pos[1] - 1] not in n_blizzard:
            new_positions.add((pos[0], pos[1] - 1))

        if (pos[0], pos[1] + 1) == last:
            print("Found last in minute:", minute)
            print("Found", (pos[0], pos[1] + 1))
            found_last = True
            break
    positions = new_positions

    if found_last is True:
        break
    

```
