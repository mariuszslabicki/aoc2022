```python
from itertools import product

def box_outside(init, min_x, min_y, min_z, max_x, max_y, max_z, obsidian, outside):
    heads = [init]
    visited = [init]
    while True:
        new_heads = []
        for head in heads:
            shifts = product([-1, 0, 1], repeat=3)
            for shift in shifts:
                if shift == [0, 0, 0]:
                    continue
                if abs(shift[0]) + abs(shift[1]) + abs(shift[2]) > 1:
                    continue
                candidate = [sum(value) for value in zip(head, shift)]
                if candidate[0] <= max_x+1 and \
                    candidate[0] >= min_x-1 and \
                    candidate[1] <= max_y+1 and \
                    candidate[1] >= min_y-1 and \
                    candidate[2] <= max_z+1 and \
                    candidate[2] >= min_z-1:

                    if candidate in outside:
                        return True, visited
                    if candidate not in visited and candidate not in obsidian:
                        new_heads.append(candidate)
                        visited.append(candidate)

        if len(new_heads) == 0:
            return False, visited
        heads = new_heads

xy_walls = dict()
yz_walls = dict()
xz_walls = dict()

cubes = []

max_x = -1
max_y = -1
max_z = -1

min_x = None
min_y = None
min_z = None

try:
    f = open("input", encoding = 'utf-8')
    for row in f:
        x, y, z = row.strip().split(",")
        x = int(x)
        y = int(y)
        z = int(z)
        if x > max_x:
            max_x = x
        if min_x is None or x < min_x:
            min_x = x
        if y > max_y:
            max_y = y
        if min_y is None or y < min_y:
            min_y = y
        if z > max_z:
            max_z = z
        if min_z is None or z < min_z:
            min_z = z
        cubes.append([x, y, z])

finally:
    f.close()

outside = [[min_x-1, min_y-1, min_z-1]]

for x in range(min_x, max_x+1):
    for y in range(min_y, max_y+1):
        for z in range(min_z, max_z+1):
            if [x, y, z] not in cubes:
                result, out_list = box_outside([x, y, z], min_x, min_y, min_z, max_x, max_y, max_z, cubes, outside)
                if result is True:
                    outside.append([x, y, z])
                    for element in out_list:
                        outside.append(element)
                else:
                    cubes.append([x, y, z])
                    for element in out_list:
                        if element not in cubes:
                            cubes.append(element)
                        

            

for cube in cubes:
    [x, y, z] = cube
    x = int(x)
    y = int(y)
    z = int(z)
    if (x, y, z) not in xy_walls:
        xy_walls[(x, y, z)] = 1
    else:
        del xy_walls[(x, y, z)]
    if (x, y, z+1) not in xy_walls:
        xy_walls[(x, y, z+1)] = 1
    else:
        del xy_walls[(x, y, z+1)]

    if (x, y, z) not in xz_walls:
        xz_walls[(x, y, z)] = 1
    else:
        del xz_walls[(x, y, z)]
    if (x, y+1, z) not in xz_walls:
        xz_walls[(x, y+1, z)] = 1
    else:
        del xz_walls[(x, y+1, z)]
        
    if (x, y, z) not in yz_walls:
        yz_walls[(x, y, z)] = 1
    else:
        del yz_walls[(x, y, z)]
    if (x+1, y, z) not in yz_walls:
        yz_walls[(x+1, y, z)] = 1
    else:
        del yz_walls[(x+1, y, z)]

disjoint = 0

for key in xy_walls:
    if xy_walls[key] == 1:
        disjoint += 1

for key in yz_walls:
    if yz_walls[key] == 1:
        disjoint += 1

for key in xz_walls:
    if xz_walls[key] == 1:
        disjoint += 1

print(disjoint)
```
