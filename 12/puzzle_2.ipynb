```python
from collections import deque

map = []
try:
    f = open("input", encoding = 'utf-8')
    one_test = []
    for row in f:
        new_line = []
        line = row.strip()
        for element in line:
            new_line.append(element)
        map.append(new_line)        

finally:
    f.close()

x_init = 0
y_init = 0

shortest_len = -1
shortest_path = []

for y in range(len(map)):
    for x in range(len(map[0])):
        if map[y][x] == 'a':
            x_init = x
            y_init = y
            paths = []
            initial_path = [(x_init, y_init)]
            paths.append(initial_path)
            print("Starting from:", initial_path)
            found_path = False

            visited_fields = [(x_init, y_init)]

            while found_path is False:
                new_paths = []
                if found_path is False:
                    lost = True
                    for path in paths:
                        last_x = path[-1][0]
                        last_y = path[-1][1]
                        fields_to_check = []
                        if map[last_y][last_x] == 'E':
                            found_len = len(path)-1
                            print("Found one path, len:", found_len)
                            if shortest_len > found_len or shortest_len == -1:
                                shortest_len = found_len
                                shortest_path = path
                            # print("Check end condition")
                            found_path = path
                            break
                        else:
                            # Type 1 - corners
                            if last_x == 0 and last_y == 0:
                                fields_to_check.append((last_x, last_y + 1))
                                fields_to_check.append((last_x + 1, last_y))
                            if last_x == len(map[0])-1 and last_y == len(map)-1:
                                fields_to_check.append((last_x, last_y - 1))
                                fields_to_check.append((last_x - 1, last_y))
                            if last_x == 0 and last_y == len(map)-1:
                                fields_to_check.append((last_x, last_y - 1))
                                fields_to_check.append((last_x + 1, last_y))
                            if last_x == len(map[0])-1 and last_y == 0:
                                fields_to_check.append((last_x, last_y + 1))
                                fields_to_check.append((last_x - 1, last_y))
                            # Type 2 - between corners
                            if last_x == 0 and last_y > 0 and last_y < len(map)-1:
                                fields_to_check.append((last_x, last_y + 1))
                                fields_to_check.append((last_x, last_y - 1))
                                fields_to_check.append((last_x + 1, last_y))
                            if last_x == len(map[0])-1 and last_y > 0 and last_y < len(map)-1:
                                fields_to_check.append((last_x, last_y + 1))
                                fields_to_check.append((last_x, last_y - 1))
                                fields_to_check.append((last_x - 1, last_y))
                            if last_x > 0 and last_x < len(map[0])-1 and last_y == 0:
                                fields_to_check.append((last_x - 1, last_y))
                                fields_to_check.append((last_x + 1, last_y))
                                fields_to_check.append((last_x, last_y + 1))
                            if last_x > 0 and last_x < len(map[0])-1 and last_y == len(map)-1:
                                fields_to_check.append((last_x - 1, last_y))
                                fields_to_check.append((last_x + 1, last_y))
                                fields_to_check.append((last_x, last_y - 1))
                            # Type 3 - area
                            if last_x > 0 and last_x < len(map[0])-1 and last_y > 0 and last_y < len(map)-1:
                                fields_to_check.append((last_x - 1, last_y))
                                fields_to_check.append((last_x + 1, last_y))
                                fields_to_check.append((last_x, last_y + 1))
                                fields_to_check.append((last_x, last_y - 1))
                            if fields_to_check == []:
                                print("Teraz nie znalazlem")
                                break

                        if map[last_y][last_x] == 'S':
                            last_height = ord('a')
                        else:
                            last_height = ord(map[last_y][last_x])
                        for field in fields_to_check:
                            if map[field[1]][field[0]] == 'E':
                                new_height = ord('z')
                            else:
                                new_height = ord(map[field[1]][field[0]])
                            if (field[0], field[1]) not in visited_fields and last_height + 1 >= new_height:
                                lost = False
                                new_path = path.copy()
                                new_path.append((field[0], field[1]))
                                new_paths.append(new_path)
                                visited_fields.append((field[0], field[1]))
                                paths = new_paths
                    if lost is True:
                        # print("Zdupilem")
                        break

print(len(shortest_path)-1)

# print(len(paths))

# for path in paths:
#     print(path[-1])
```
