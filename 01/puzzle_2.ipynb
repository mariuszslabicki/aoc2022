```python
try:
   f = open("input", encoding = 'utf-8')
   sums = []
   current_sum = 0
   for row in f:
    if len(row) == 1:
        sums.append(current_sum)
        current_sum = 0
    else:
        current_sum += int(row)

finally:
    sums.append(current_sum)
    f.close()

sums.sort(reverse=True)

main_sum = sums[0] + sums[1] + sums[2]

print(main_sum)
```
