```python

from dataclasses import dataclass

@dataclass
class Elf:
    x: int
    y: int
    next_x: int = None
    next_y: int = None

elves = []
directions = ['N', 'S', 'W', 'E']

try:
    f = open("input", encoding = 'utf-8')
    y = 0
    for row in f:
        for x in range(len(row)):
            if row[x] == '#':
                elves.append(Elf(x, y))
        y += 1

finally:
    f.close()

# for elf in elves:
#     print(elf)

for round in range(10):
    # print("Runda", round)
    # print(directions)
    for elf in elves:
        ignore_this_elf = True
        for iterator in elves:
            if elf.x - 1 == iterator.x and elf.y - 1 == iterator.y or \
                elf.x == iterator.x and elf.y - 1 == iterator.y or \
                elf.x + 1 == iterator.x and elf.y - 1 == iterator.y or \
                elf.x - 1 == iterator.x and elf.y + 1 == iterator.y or \
                elf.x == iterator.x and elf.y + 1 == iterator.y or \
                elf.x + 1 == iterator.x and elf.y + 1 == iterator.y or \
                elf.x - 1 == iterator.x and elf.y == iterator.y or \
                elf.x + 1 == iterator.x and elf.y == iterator.y:
                ignore_this_elf = False
                break
        if ignore_this_elf is True:
            continue
        found_new_point = False
        for dir in directions:
            colision = False
            if dir == 'N':
                for iterator in elves:
                    if elf.x - 1 == iterator.x and elf.y - 1 == iterator.y or \
                        elf.x == iterator.x and elf.y - 1 == iterator.y or \
                        elf.x + 1 == iterator.x and elf.y - 1 == iterator.y:
                        colision = True
                        break
                if colision is True:
                    continue
                elf.next_x = elf.x
                elf.next_y = elf.y - 1
                break
            if dir == 'S':
                for iterator in elves:
                    if elf.x - 1 == iterator.x and elf.y + 1 == iterator.y or \
                        elf.x == iterator.x and elf.y + 1 == iterator.y or \
                        elf.x + 1 == iterator.x and elf.y + 1 == iterator.y:
                        colision = True
                        break
                if colision is True:
                    continue
                elf.next_x = elf.x
                elf.next_y = elf.y + 1
                break
            if dir == 'W':
                for iterator in elves:
                    if elf.x - 1 == iterator.x and elf.y - 1 == iterator.y or \
                        elf.x - 1 == iterator.x and elf.y == iterator.y or \
                        elf.x - 1 == iterator.x and elf.y + 1 == iterator.y:
                        colision = True
                        break
                if colision is True:
                    continue
                elf.next_x = elf.x - 1
                elf.next_y = elf.y
                break
            if dir == 'E':
                for iterator in elves:
                    if elf.x + 1 == iterator.x and elf.y - 1 == iterator.y or \
                        elf.x + 1 == iterator.x and elf.y == iterator.y or \
                        elf.x + 1 == iterator.x and elf.y + 1 == iterator.y:
                        colision = True
                        break
                if colision is True:
                    continue
                elf.next_x = elf.x + 1
                elf.next_y = elf.y
                break
            if found_new_point is True:
                continue
    # print("Proposals")
    # for elf in elves:
    #     print(elf)
    new_positions = {}
    for elf in elves:
        if elf.next_x is not None and elf.next_y is not None:
            if (elf.next_x, elf.next_y) not in new_positions:
                new_positions[(elf.next_x, elf.next_y)] = 1
            else:
                new_positions[(elf.next_x, elf.next_y)] = new_positions[(elf.next_x, elf.next_y)] + 1

    for elf in elves:
        if elf.next_x is not None and elf.next_y is not None:
            if new_positions[(elf.next_x, elf.next_y)] == 1:
                elf.x = elf.next_x
                elf.y = elf.next_y
        elf.next_x = None
        elf.next_y = None

    direction = directions.pop(0)
    directions.append(direction)
    # print("After round", round)
    # for elf in elves:
    #     print(elf)

min_x = None
max_x = None
min_y = None
max_y = None

for elf in elves:
    if min_x is None:
        min_x = elf.x
    if max_x is None:
        max_x = elf.x
    if min_y is None:
        min_y = elf.y
    if max_y is None:
        max_y = elf.y

    if elf.x < min_x:
        min_x = elf.x
    if elf.x > max_x:
        max_x = elf.x
    if elf.y < min_y:
        min_y = elf.y
    if elf.y > max_y:
        max_y = elf.y

# print(min_x, max_x)
# print(min_y, max_y)

area = (max_x - min_x + 1) * (max_y - min_y + 1)
# print(area)

result = area - len(elves)
print(result)
```
