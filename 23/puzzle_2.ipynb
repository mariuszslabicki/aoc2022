```python
# %%timeit

# 7.88 ms

# Results
# * 500 is too low
# * 5000 is too high
# * 2000 is too high

import time
from dataclasses import dataclass
from collections import deque

@dataclass
class Elf:
    x: int
    y: int
    N_free: bool = True
    S_free: bool = True
    E_free: bool = True
    W_free: bool = True
    next_x: int = None
    next_y: int = None

elves = []
# directions = ['N', 'S', 'W', 'E']
directions = [1, 2, 3, 4]
# directions = deque(directions)

locations = set()

try:
    f = open("input", encoding = 'utf-8')
    y = 0
    for row in f:
        for x in range(len(row)):
            if row[x] == '#':
                elves.append(Elf(x, y))
                locations.add((x, y))
        y += 1

finally:
    f.close()

# for elf in elves:
#     print(elf)

print("Elves:", len(elves))

round = 0
# while round < 10:
while True:
    start = time.time()
    # print("Round", round)
    # print(directions)
    number_of_sleeping_elves = 0
    new_positions = {}
    for elf in elves:
        if (elf.x + 1, elf.y) not in locations and \
            (elf.x - 1, elf.y) not in locations and \
            (elf.x + 1, elf.y - 1) not in locations and \
            (elf.x, elf.y - 1) not in locations and \
            (elf.x - 1, elf.y - 1) not in locations and \
            (elf.x + 1, elf.y + 1) not in locations and \
            (elf.x, elf.y + 1) not in locations and \
            (elf.x - 1, elf.y + 1) not in locations:
            continue
        if elf.N_free or elf.S_free or elf.W_free or elf.E_free:
            for iterator in elves:
                if elf == iterator:
                    continue
                x_dist = abs(elf.x - iterator.x)
                y_dist = abs(elf.y - iterator.y)
                if x_dist <= 1 and y_dist <= 1:
                    if iterator.y < elf.y:
                        elf.N_free = False
                        iterator.S_free = False
                    if iterator.y > elf.y:
                        elf.S_free = False
                        iterator.N_free = False
                    if iterator.x < elf.x:
                        elf.W_free = False
                        iterator.E_free = False
                    if iterator.x > elf.x:
                        elf.E_free = False
                        iterator.W_free = False
                    if not elf.N_free and not elf.S_free and not elf.W_free and not elf.E_free:
                        break

        if elf.N_free and elf.S_free and elf.W_free and elf.E_free:
            continue
        if not elf.N_free and not elf.S_free and not elf.W_free and not elf.E_free:
            continue

        if not elf.N_free or not elf.S_free or not elf.W_free or not elf.E_free:
            for direction in directions:
                if direction == 1:
                    if elf.N_free:
                        elf.next_x = elf.x
                        elf.next_y = elf.y - 1
                        break
                if direction == 2:
                    if elf.S_free:
                        elf.next_x = elf.x
                        elf.next_y = elf.y + 1
                        break
                if direction == 3:
                    if elf.W_free:
                        elf.next_x = elf.x - 1
                        elf.next_y = elf.y
                        break
                if direction == 4:
                    if elf.E_free:
                        elf.next_x = elf.x + 1
                        elf.next_y = elf.y
                        break

    
        if elf.next_x is not None and elf.next_y is not None:
            if (elf.next_x, elf.next_y) not in new_positions:
                new_positions[(elf.next_x, elf.next_y)] = 1
            else:
                new_positions[(elf.next_x, elf.next_y)] = new_positions[(elf.next_x, elf.next_y)] + 1
                elf.next_x = None
                elf.next_y = None
    locations = set()
    for elf in elves:
        if elf.next_x is not None and elf.next_y is not None:
            if new_positions[(elf.next_x, elf.next_y)] == 1:
                elf.x = elf.next_x
                elf.y = elf.next_y

        if elf.N_free and elf.S_free and elf.W_free and elf.E_free or \
            not elf.N_free and not elf.S_free and not elf.W_free and not elf.E_free:
            number_of_sleeping_elves += 1
        locations.add((elf.x, elf.y))
        elf.next_x = None
        elf.next_y = None
        elf.N_free = True
        elf.S_free = True
        elf.W_free = True
        elf.E_free = True


    direction = directions.pop(0)
    directions.append(direction)
    # direction = directions.popleft()
    # directions.append(direction)

    round += 1
    # print("Sleeping:", number_of_sleeping_elves)
    end = time.time()
    print("Finished round", round, "took:", end-start, "sleeping", number_of_sleeping_elves)
    if number_of_sleeping_elves == len(elves):
        print("Koniec")
        print("W rundzie", round)
        break
    # for elf in elves:
    #     print(elf)

# min_x = None
# max_x = None
# min_y = None
# max_y = None

# for elf in elves:
#     if min_x is None:
#         min_x = elf.x
#     if max_x is None:
#         max_x = elf.x
#     if min_y is None:
#         min_y = elf.y
#     if max_y is None:
#         max_y = elf.y

#     if elf.x < min_x:
#         min_x = elf.x
#     if elf.x > max_x:
#         max_x = elf.x
#     if elf.y < min_y:
#         min_y = elf.y
#     if elf.y > max_y:
#         max_y = elf.y

# # print(min_x, max_x)
# # print(min_y, max_y)

# area = (max_x - min_x + 1) * (max_y - min_y + 1)
# # print(area)

# result = area - len(elves)
# print(result)
```
